```python
# Import required libraries and dependencies
import pandas as pd
import hvplot
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import panel as pn

import os
import warnings
warnings.simplefilter("ignore")
os.environ["PYTHONWARNINGS"] = "ignore"
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
cryptocurrencies_market_data = df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
# Display plot

pane = pn.panel(cryptocurrencies_market_data)
pane
```

```python
#Save the plot to the output folder

hvplot.save(cryptocurrencies_market_data,"output/cryptocurrencies_market_data_plot.html")
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
scaler = StandardScaler()
crypto_scaled = scaler.fit_transform(df_market_data)
crypto_scaled[0:5]
```

```python
# Create a DataFrame with the scaled data
df_market_data_scaled = pd.DataFrame(crypto_scaled, columns=df_market_data.columns)
df_market_data_scaled.head()
```

```python
# Create a DataFrame with the scaled data
df_market_data_scaled = pd.DataFrame(crypto_scaled, columns=df_market_data.columns)

# Copy the 'coin_id' column from the original data to the scaled DataFrame
df_market_data_scaled["coin_id"] = df_market_data.index

# Set the 'coin_id' column as index
df_market_data_scaled.set_index('coin_id', inplace=True) 

# Display sample data
df_market_data_scaled.head()
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1,11))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list
for i in k:
    k_model = KMeans(n_clusters=i, random_state=1, n_init=10)
    k_model.fit(df_market_data_scaled)
    inertia.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k, "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve
df1_elbow = pd.DataFrame(elbow_data)
df1_elbow.head()
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
df1_elbow_curve = df1_elbow.hvplot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k
)
#Save the plot to the output folder
hvplot.save(df1_elbow_curve,"output/df1_elbow_curve_plot.html")

# Display plot
df1_elbow_curve
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** The best value for k is 4.

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
k_model = KMeans(n_clusters=4, random_state=1)
```

```python
# Fit the K-Means model using the scaled data
k_model.fit(df_market_data_scaled)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
crypto_clusters = k_model.predict(df_market_data_scaled)

# Print the resulting array of cluster values.
crypto_clusters
```

```python
# Create a copy of the DataFrame
df_martket_scaled_predictions = df_market_data_scaled.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
df_martket_scaled_predictions["predicted_clusters"] = crypto_clusters

# Display sample data
df_martket_scaled_predictions.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
crypto_clusters_scatter = df_martket_scaled_predictions.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by= "predicted_clusters",
    hover_cols= ["coin_id","price_change_percentage_24h", "price_change_percentage_7d"],
    title="Cryptocurrency Price Changes (24h vs. 7d)"
)
#Save the plot to the output folder
hvplot.save(crypto_clusters_scatter,"output/crypto_clusters_scatter_plot.html")

# Display plot
crypto_clusters_scatter
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca =PCA(n_components=3)
pca
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
clusters_pca = pca.fit_transform(df_martket_scaled_predictions)
# View the first five rows of the DataFrame. 
clusters_pca[:5]
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
pca.explained_variance_ratio_
```

```python
#calculate the Total Explained Variance by summing all 3 Explained Variance Ratios
sum(pca.explained_variance_ratio_)
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** About 88% of the total variance is condensed into the 3 PCA variables.

```python
# Create a new DataFrame with the PCA data.
clusters_pca_df = pd.DataFrame(
    clusters_pca,
    columns=["PCA1", "PCA2","PCA3"]
)
# Copy the crypto names from the original data
clusters_pca_df["coin_id"] = df_market_data.index

# Set the coinid column as index
clusters_pca_df.set_index('coin_id', inplace=True) 

# Display sample data
clusters_pca_df.head()
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k_pca = list(range(1,11))
```

```python

# Create an empty list to store the inertia values
inertia2 = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list

for i in k_pca:
    k_model = KMeans(n_clusters=i, random_state=1, n_init=10)
    k_model.fit(clusters_pca_df)
    inertia2.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_pca_data = {"k_pca":k_pca, "inertia2":inertia2}
# Create a DataFrame with the data to plot the Elbow curve
df2_elbow = pd.DataFrame(elbow_pca_data)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
df2_elbow_curve = df2_elbow.hvplot.line(
    x="k_pca", 
    y="inertia2", 
    title="Elbow Curve", 
    xticks=k
)
#Save the plot to the output folder
hvplot.save(df2_elbow_curve ,"output/df2_elbow_curve_plot.html")

# Display the plot
df2_elbow_curve
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** The best value for k when using PCA data is 4.


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** No, the k value has not changed from the what i found using the origninal data. 

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
crypto_model = KMeans(n_clusters=4, random_state=1, n_init=10)
crypto_model
```

```python
# Fit the K-Means model using the PCA data
k_model.fit(clusters_pca_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
crypto2_clusters = k_model.predict(clusters_pca_df)
# Print the resulting array of cluster values.
crypto2_clusters
```

```python
# Create a copy of the DataFrame with the PCA data
predicted_clusters_pca_df = clusters_pca_df.copy()

# Add a new column to the DataFrame with the predicted clusters
predicted_clusters_pca_df["pca_prediction"] = crypto2_clusters

# Display sample data
predicted_clusters_pca_df.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
crypto2_clusters_scatter = predicted_clusters_pca_df.hvplot.scatter(
    x="PCA1",
    y="PCA2",
    by= "pca_prediction",
    hover_cols= ["coin_id","PCA1", "PCA2"],
    title="Cryptocurrency PCA Analysis (PC1 vs. PC2)"
)
#Save the plot to the output folder
hvplot.save(crypto2_clusters_scatter,"output/crypto2_clusters_scatter_plot.html")

# Display plot
crypto2_clusters_scatter
```
